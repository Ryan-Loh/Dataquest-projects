## Intro:
Working with the Titanic dataset from Kaggle.com, The goal of this project is to generate conditional plots and practice using the seaborn toolset

The dataset "train.csv" contains several columns:
1. PassengerId -- A numerical id assigned to each passenger.
2. Survived -- Whether the passenger survived (1), or didn't (0).
3. Pclass -- The class the passenger was in.
4. Name -- the name of the passenger.
5. Sex -- The gender of the passenger -- male or female.
6. Age -- The age of the passenger. Fractional.
7. SibSp -- The number of siblings and spouses the passenger had on board.
8. Parch -- The number of parents and children the passenger had on board.
9. Ticket -- The ticket number of the passenger.
10. Fare -- How much the passenger paid for the ticket.
11. Cabin -- Which cabin the passenger was in.
12. Embarked -- Where the passenger boarded the Titanic.

```python
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Importing data and removing unimportant columns of "ticket", and "name", also deleting NaN rows
titanic = pd.read_csv("train.csv")
titanic = titanic[["Survived", "Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]]
titanic = titanic.dropna()
```

```python
titanic.head()
```

```python
sns.distplot(titanic["Fare"])
```

```python
sns.distplot(titanic["Age"])
```

```python
sns.kdeplot(titanic["Age"],shade=True)
plt.xlabel("Age")
```

```python
sns.set_style("dark")
sns.kdeplot(titanic["Age"],shade=True)
sns.despine(left=True, bottom=True)
plt.xlabel("Age")
```

```python
g = sns.FacetGrid(titanic, col = "Pclass", height = 5)
g.map(sns.kdeplot,"Age",shade=True)
sns.despine(left=True, bottom=True)
```

```python
g = sns.FacetGrid(titanic, col="Survived", row="Pclass", height = 5)
g.map(sns.kdeplot,"Age",shade=True)
```

```python
g = sns.FacetGrid(titanic, col = "Survived", row = "Pclass", hue = "Sex", height = 3)
g.map(sns.kdeplot, "Age", shade = True).add_legend()
sns.despine(left=True, bottom=True)
```
